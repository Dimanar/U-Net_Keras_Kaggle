```python
import os
import cv2
import matplotlib.pyplot as plt
import numpy as np
import skimage
import seaborn as sns
from tqdm import tqdm
import pandas as pd
```

```python
PATH_TRAIN = 'data/train/'
PATH_TEST = 'data/test/'
```

```python
(image, mask) = (np.load('data/x_train.npy', allow_pickle=True),
                np.load('data/y_train.npy', allow_pickle=True))
```

```python
train_ids = next(os.walk(PATH_TRAIN))[1]
```

```python
def get_shapes(img, shp=[]):
    for img_ in img:
        shp.append(img_.shape)
    return shp
```

```python
data = pd.DataFrame({
    'images': list(image),
    'masks': list(mask),
    'images_shape': get_shapes(image, shp=list()),
    'masks_shape': get_shapes(mask, shp=list()),
    'names': train_ids
})
```

```python
data.head(3)
```

```python
msk_unique_shapes = data.masks_shape.unique()
img_unique_shapes = data.images_shape.unique()
```

```python
def to_str(array):
    return "{}x{} size".format(str(array[0]), str(array[1]))
```

```python
def show_image(df, colum, unique_shape, x, y=4, seed=42):
    fig, axs = plt.subplots(x, y)
    fig.set_figwidth(18)    
    fig.set_figheight(36)
    
    for i in range(x):
        shape_ = unique_shape[i]
        
        selection = df[(df[ colum[0] ] == shape_)].sample(
            n=y, replace=True, random_state=seed)
        img_ = selection[colum[1]].values
        img_ = iter(img_)

        for j in range(y):
            axs[i, j].imshow(next(img_))
            axs[i, j].set_title(to_str(shape_))
            
    plt.show()
```

```python
show_image(df=data, unique_shape=img_unique_shapes, colum=['images_shape', 'images'],
          x=9, y=4, seed=47)
```

```python
show_image(df=data, unique_shape=msk_unique_shapes, colum=['masks_shape', 'masks'],
          x=len(msk_unique_shapes), y=4, seed=47)
```

```python
def del_d(series):
    func = lambda x: str((x[0], x[1]))
    return list(map(func, series))
```

```python
def reshape(original, to_resize):
    x, y = original.shape[:2]
    img = resize(to_resize, (x, y), mode='constant')
    return img
```

```python
def show_result(test, pred):
    fig, axs = plt.subplots(1, 2)
    n = np.random.randint(len(test))
    axs[0].imshow(test[n], cmap='gray')
    axs[0].set_title('Original')
    axs[1].imshow(reshape(test[n], pred[n]), cmap='gray')
    axs[1].set_title('Predicted')
    plt.show()
```
